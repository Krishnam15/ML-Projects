```python
import pandas as pd
import numpy as np
import ast
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
credits.sample(3)
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head(2)
```

```python
movies.info()
```

```python
# Columns needed for movie recommendation
# (Ignoring all numerical columns for this)
# genres
# id (needed later to display movie posters in recommendation system)
# keywords
# title (not original_title as it can vary with the language of movie)
# overview
# cast (recommending on the basis of actor)
# crew (recommending on the basis of director)
```

```python
movies = movies[['genres','id','keywords','title','overview','cast','crew']]
```

```python
movies.head(2)
```

```python
movies.iloc[0].genres
```

```python
def convert(obj): # for converting genres and keywords in clean format.
    list = []
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list   
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(5)
```

```python
def fetch_top3(obj): # for converting cast in clean format with selecting top 3 cast members only.
    list = []
    counter = 0
    for i in ast.literal_eval(obj):
        if (counter<3):
            list.append(i['name'])
        counter=counter + 1;    
    return list    
```

```python
movies['cast'] = movies['cast'].apply(fetch_top3)
```

```python
movies.sample(3)
```

```python
def fetch_director(obj): # to get the director name.
    list = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            list.append(i['name'])
            break
    return list
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(3)
```

```python
def convert_ov(obj): # converting the overview from str to list. 
    list = []
    list.append(obj)
    return list
```

```python
movies['overview'] = movies['overview'].apply(convert_ov)
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head(3)
```

```python
movies['tags'].iloc[0]
```

```python
new_df = movies[['id','title','tags']]
new_df.head(3)
```

```python
# def convert_to_str(obj):
#     str = " ".join(obj)
#     return str
```

```python
# new_df['tags'] = new_df['tags'].apply(str)
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x: " ".join(str(item) for item in x)) # list to string conversion of tags column
```

```python
new_df.head()
```

```python
type(new_df['tags'].iloc[0])
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.sample(3) 
```

```python
import nltk # to transform similar words like 'action' and 'actions' into one
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer() 
#PorterStemmer is a stemming algorithm from the Natural Language Toolkit (NLTK) in Python. 
# It reduces words to their root form (stem) by removing suffixes. 
# This helps in text normalization and improves search and text analysis.
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
# VECTORIZATION - Selecting 5000 most repeating words and forming their vector  
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
# list(enumerate(similarity[0]))
```

```python
def recommend_movie(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend_movie('Batman Begins')
```
